```python
from churn_gym.infrastructure.logging.config import configure_logging
from churn_gym.infrastructure.logging.loguru_logger import LoguruLogger

# Configuration DU logging (UNE SEULE FOIS, point d'entrée)
configure_logging(env="dev")

# Logger instancié UNE FOIS (pas à chaque appel)
logger = LoguruLogger()


def compute(a: float, b: float) -> float:
    """
    Calcule (a + b)² avec journalisation structurée.
    """

    logger.debug(
        "Starting computation",
        a=a,
        b=b,
        operation="(a + b) ** 2",
    )

    result = (a + b) ** 2

    logger.info(
        "Computation finished",
        a=a,
        b=b,
        result=result,
    )

    return result

# ── Essais concrets
res1 = compute(2, 3)
print("Result 1 =", res1)

res2 = compute(5.5, 1.5)
print("Result 2 =", res2)

res3 = compute(-1, 4)
print("Result 3 =", res3)
```
